```python
# Function to check if a number is prime
def is_prime(num):
    if num <= 1:
        return False

    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False

    return True

# Input from the user
number = int(input("Enter a number: "))

# Check and display the result
if is_prime(number):
    print(f"{number} is a prime number.")
else:
    print(f"{number} is not a prime number.")
    
```

```python
import random

# Generate two random numbers between 1 and 10
num1 = random.randint(1, 10)
num2 = random.randint(1, 10)

# Ask the user for the product
answer = int(input(f"What is {num1} × {num2}? "))

# Check the answer
if answer == num1 * num2:
    print("Correct! Well done.")
else:
    print(f"Incorrect. The correct answer is {num1 * num2}.")
```

```python
# This program prints the squares of all even numbers
# between 100 and 200.

for num in range(100, 201):
    if num % 2 == 0:
        print(num, ":", num * num)
```

```python
# Hospital Billing System

room_charges = 4 * 2000
doctor_fee = 1500
lab_charges = 3 * 300
medicine_charges = 2400

total_bill = room_charges + doctor_fee + lab_charges + medicine_charges

if total_bill > 10000:
    discount = total_bill * 0.10
else:
    discount = 0

final_bill = total_bill - discount

print("Room Charges =", room_charges)
print("Doctor Fee =", doctor_fee)
print("Lab Charges =", lab_charges)
print("Medicine Charges =", medicine_charges)
print("Total Bill =", total_bill)
print("Discount =", discount)
print("Final Bill =", final_bill)
```
