Problem Statement : Hospital Patient Data Analysis

1. Load the patient dataset and show summary with info()

```python
import pandas as pd

patient_df = pd.read_csv("Patient_Data.csv")
billing_df = pd.read_csv("Billing_Data.csv")

print(patient_df.info())
```

2.Select only billing-related columns

```python
billing_columns = patient_df[['PatientID', 'Department', 'Doctor', 'BillAmount']]
print(billing_columns)
```

3.Drop administrative columns

```python
patient_clean = patient_df.drop(
    columns=['ReceptionistID', 'CheckInTime']
)

print(patient_clean.head())
```

4.Total bill amount per department using groupby()

```python
dept_bill = patient_df.groupby('Department')['BillAmount'].sum()
print(dept_bill)
```

5.Remove duplicate patient records

```python
patient_nodup = patient_df.drop_duplicates(subset='PatientID')
print(patient_nodup)
```

6.Fill missing BillAmount values with mean

```python
mean_bill = patient_df['BillAmount'].mean()

patient_df['BillAmount'] = patient_df['BillAmount'].fillna(mean_bill)

print(patient_df)
```

7.Merge billing dataset with patient dataset

```python
merged_df = pd.merge(
    patient_df,
    billing_df,
    on='PatientID',
    how='inner'
)

print(merged_df)
```

8.Concatenate new patient records (row-wise)

```python
new_patients = pd.DataFrame({
    'PatientID':[106,107],
    'Name':['John','Sara'],
    'Department':['Cardiology','Neurology'],
    'Doctor':['Dr. Patel','Dr. Rao'],
    'BillAmount':[6500,7200],
    'ReceptionistID':[4,5],
    'CheckInTime':['2023-01-15 09:00','2023-01-15 10:00']
})

updated_patients = pd.concat(
    [patient_df, new_patients],
    ignore_index=True
)

print(updated_patients)
```

9.Concatenate new billing columns (column-wise)

```python
new_billing_cols = billing_df[['InsuranceCovered', 'FinalAmount']]

final_dataset = pd.concat(
    [patient_df, new_billing_cols],
    axis=1
)

print(final_dataset)
```

10.Conclusion 

The hospital patient dataset was successfully cleaned and processed using Pandas. Missing bill amounts were replaced with the mean value, duplicate patient records were removed, and department-wise revenue was calculated using groupby operations. The patient dataset was merged with billing information through PatientID, and new patient records and billing columns were added using concatenation. The final dataset is complete, consistent, and suitable for further analysis such as department revenue evaluation and doctor performance assessment.
