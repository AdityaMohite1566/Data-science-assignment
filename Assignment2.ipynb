```python
# List operations

numbers = [10, 20, 30, 40, 50]
print("Original List:", numbers)

# append
numbers.append(60)
print("After append:", numbers)

# extend
numbers.extend([70, 80])
print("After extend:", numbers)

# insert
numbers.insert(2, 25)
print("After insert:", numbers)

# remove
numbers.remove(40)
print("After remove:", numbers)

# pop
numbers.pop()
print("After pop:", numbers)

# index
print("Index of 30:", numbers.index(30))

# count
print("Count of 20:", numbers.count(20))

# sort
numbers.sort()
print("After sort:", numbers)

# reverse
numbers.reverse()
print("After reverse:", numbers)

# clear
numbers.clear()
print("After clear:", numbers)
```

```python
# Dictionary operations

person = {
    "name": "Aditya",
    "age": 22,
    "city": "Satara"
}

print("Original Dictionary:")
print(person)

```

### Section c:Exploring Pandas

```python
import pandas as pd
import numpy as np

# Create DataFrame
data = {
    "Name": ["Aditya", "Rahul", "Priya", "Sneha", "Amit",
             "Kiran", "Rohit", "Pooja", "Neha", "Aditya"],
    
    "Age": [22, 23, 21, 24, 25,
            22, 23, 24, 21, 22],
    
    "City": ["Satara", "Pune", "Mumbai", "Kolhapur", "Nashik",
             "Pune", "Mumbai", "Satara", "Kolhapur", "Satara"],
    
    "Salary": [30000, 35000, np.nan, 40000, 45000,
               38000, 35000, np.nan, 42000, 30000]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

# 1. info()
print("\nDataFrame Info:")
df.info()

# 2. describe()
print("\nDataFrame Description:")
print(df.describe())

# 3. Select Name and City columns
print("\nName and City Columns:")
print(df[["Name", "City"]])

# 4. Drop City column
df_drop = df.drop("City", axis=1)
print("\nAfter Dropping City Column:")
print(df_drop)

# 5. Fill missing Salary values with mean
mean_salary = df["Salary"].mean()
df["Salary"] = df["Salary"].fillna(mean_salary)

print("\nAfter Filling Missing Salary Values:")
print(df)

# 6. Remove duplicate rows
df = df.drop_duplicates()

print("\nAfter Removing Duplicate Rows:")
print(df)
```
