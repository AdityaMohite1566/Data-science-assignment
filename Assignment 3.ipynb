##Descriptive Analytics and Data Preprocessing on Sales & Discounts Dataset

```python
# Import libraries
import pandas as pd

# Load dataset
df = pd.read_csv("sales_data_with_discounts.csv")

# Display first 5 rows
print(df.head())

# Check dataset information
print("\nDataset Information:")
print(df.info())

# Identify numerical columns
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns

print("\nNumerical Columns:")
print(numerical_columns)
```

```python


numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns

for col in numerical_columns:
    print(f"\n{col}")
    print("Mean:", df[col].mean())
    print("Median:", df[col].median())
    print("Mode:", df[col].mode().tolist())
    print("Standard Deviation:", df[col].std())
```

### Interpretation of Descriptive Statistics

#### 1. Volume

The mean volume is 5.07, the median is 4, and the mode is 3. Since the mean is slightly higher than the median, the distribution is positively skewed. The standard deviation of 4.23 indicates moderate variation in the sales volume.

#### 2. Average Price

The mean average price is 10453.43, while the median is only 1450. The large difference between the mean and median indicates a highly positively skewed distribution. This suggests that a few products have very high prices compared to the majority of products. The high standard deviation of 18079.90 confirms significant variability in product prices.

#### 3. Total Sales Value

The mean total sales value is 33812.84, whereas the median is 5700. Since the mean is much greater than the median, the distribution is positively skewed. A few transactions have exceptionally high sales values, which increase the average. The high standard deviation of 50535.07 indicates a wide spread in sales values.

#### 4. Discount Rate (%)

The mean discount rate is 15.16%, and the median is 16.58%. These values are relatively close, indicating that discount rates are fairly consistent across transactions. The standard deviation of 4.22 shows moderate variation in discount percentages.

#### 5. Discount Amount

The mean discount amount is 3346.50, while the median is 988.93. Since the mean is considerably higher than the median, the distribution is positively skewed. This indicates that some transactions received very large discounts. The standard deviation of 4509.90 shows substantial variation in discount amounts.

#### 6. Net Sales Value

The mean net sales value is 30466.34, while the median is 4677.79. The large difference between the mean and median suggests a positively skewed distribution. A few high-value transactions contribute significantly to the overall sales revenue. The high standard deviation indicates considerable variability in net sales values.

```python
# Select numerical columns
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns

# Create box plots
for col in numerical_columns:
    plt.figure(figsize=(6,4))
    plt.boxplot(df[col])
    plt.title(f'Box Plot of {col}')
    plt.ylabel(col)
    plt.show()
```

```python
# Select categorical columns
categorical_columns = df.select_dtypes(include=['object']).columns

print("Categorical Columns:")
print(categorical_columns)

# Create bar charts
for col in categorical_columns:
    plt.figure(figsize=(8,5))
    df[col].value_counts().plot(kind='bar')
    plt.title(f'Bar Chart of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
```

Bar Chart Analysis

The bar charts show the frequency distribution of categorical variables.

Some categories occur much more frequently than others.

Categories with taller bars represent the most common observations in the dataset.

Categories with shorter bars indicate less frequent occurrences.

Insights:
